```python
with open('../tulane_images/source/AbsolutePixelCoordinate.txt', 'r') as f:
    lines = f.readlines()

items = [x.strip().split(' ') for x in lines]

from ast import literal_eval

items = [(x[0].split('.')[0], literal_eval(x[1])) for x in items]
items.sort(key=lambda x: int(x[0]))
# items = items[:10]
items
```

```python
source_list = []
min_y = min([item[1][1] for item in items])
for item in items:
    file_name = f'{item[0]}.tiff'
    source_list.append({
        'path': f'./{file_name}',
        'z': 0,
        'position': {
            'x': round(item[1][0]),
            'y': round(item[1][1] - min_y),
            'scale': 1
        },
        'params': {'style': {'bands': [
            {'palette': '#f00', 'band': 1},
            {'palette': '#0f0', 'band': 2},
            {'palette': '#00f', 'band': 3},
            {
                'palette': ['#fff0', '#ffff'],
                'band': 4,
                'composite': 'multiply'
            }
        ]}}
    })
len(source_list), source_list[:2]
```

```python
time_series = []
# time_series = source_list.copy() # prepend complete composite

step_size = 5
for i, f in enumerate(range(0, len(source_list), step_size)):
    for s in source_list[:f+1]:
        s = {**s}
        s['z'] = i + 1
        time_series.append(s)

len(time_series), time_series[-3:]
```

```python
import yaml
yaml.Dumper.ignore_aliases = lambda *args : True
yaml_str = f'---\n{yaml.dump({"sources": time_series})}'

with open('../tulane_images/masked/timestep.yaml', 'w') as f:
    f.write(yaml_str)
```
