```python
import os
source_files = os.listdir('../tulane_images/source')
source_files = [f.split('.')[0] for f in source_files if f.endswith('.Raw')]
source_files[:10]
```

```python
# read header (generated to interpret raw data of our specified size (6464, 4852))
# this header was generated by David Manthey with tifftools
with open('./header.bin', 'rb') as f:
    header = f.read()
header
```

```python
for source_file in source_files:
    with open(f'../tulane_images/source/{source_file}.Raw', 'rb') as f:
        data = f.read()
    with open(f'../tulane_images/rawtiff/{source_file}.tiff', 'wb') as f:
        f.write(header + data)
    print(f'Added header to {source_file}')
```

```python
import rawpy
import imageio

for source_file in source_files:
    with rawpy.imread(f'../tulane_images/rawtiff/{source_file}.tiff') as raw:
        rgb = raw.postprocess(gamma=(1,1))
        imageio.imsave(f'../tulane_images/libraw_demosaic/{source_file}.tiff', rgb)
    print(f'Demosaic complete for {source_file}')
```

```python
# clean up temp files
os.system('rm ../tulane_images/rawtiff/*.tiff')
```
