```python
import os
files = os.listdir('../tulane_images/output')
```

```python
timestamps = [file.split('.')[0] for file in files if 'Raw' in file]
timestamps = sorted(timestamps, key=int)
timestamps = [(t, (t[:4], t[4:6], t[6:8], t[8:10], t[10:12], t[12:14], t[14:])) for t in timestamps]
timestamps = [(ts[0], tuple(int(t) for t in ts[1])) for ts in timestamps]
timestamps
```

```python
deltas = [(timestamps[i+1][1][-1] - ts[1][-1]) for i, ts in enumerate(timestamps[:-1])]
deltas = [(d + 1000) % 1000 for d in deltas]
deltas
```

```python
min(deltas), max(deltas), sum(deltas) / len(deltas)
```

```python
import math

ms_per_frame = 1000/60
delta_scale = ms_per_frame / min(deltas)
delta_scale = math.ceil(2 * delta_scale) # for safety
print(ms_per_frame)
print(delta_scale)

scaled_ts = [(ts[0], 1000*ts[1][-2] + ts[1][-1]) for ts in timestamps]
scaled_ts = [(ts[0], delta_scale * ts[1]) for ts in scaled_ts]
scaled_ts = [(ts[0], ts[1] - scaled_ts[0][1]) for ts in scaled_ts]
scaled_ts
```

```python
timescale = 1
ms_duration = scaled_ts[-1][1]

frame_list = []
for ms in range(0, ms_duration, int(timescale * ms_per_frame)):
    for i, ts in enumerate(scaled_ts):
        ts_next = scaled_ts[i + 1]
        if ts_next[1] >= ms:
            frame_list.append(ts[0])
            break

frame_list
```

```python
len(frame_list) / 60
```

```python
for i, frame in enumerate(frame_list):
    # os.system(f'ln -s ../png_output/{frame}.Raw.png tulane_images/video/frame_{i+1:04d}.png')
    os.system(f'ln -s ../output/{frame}.Raw.tiff ../tulane_images/video/frame_{i+1:04d}.tiff')
    if i % 100 == 0:
        print(i)
```
